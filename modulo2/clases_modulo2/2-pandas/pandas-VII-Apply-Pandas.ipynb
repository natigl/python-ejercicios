## `apply()` 

### 쯈u칠 es un `apply`? 

En el caso de que necesitemos aplicar una misma funci칩n a las filas o columnas de un objeto DataFrame de Pandas contamos con el m칠todo `apply()`. Con 칠l es posible aplicar funciones (tanto creadas por nosotras como las *built-in functions* de Python) a los conjuntos de datos. Esto hace que sea un m칠todo interesante de conocer y, cuando se comprende bien su funcionamiento, permite realizar operaciones complejas de forma eficiente. Por eso en esta entrada vamos a ver c칩mo aplicar una funci칩n a las filas o columnas de un DataFrame con `apply()`. 

La funci칩n `apply` de Pandas DataFrame toma una funci칩n como argumento y la aplica a lo largo de un eje del DataFrame.

Basicamente lo que va a hacer este m칠todo es ir celda a celda de la columna que especifiquemos y le aplicara a los valores de la celda la funci칩n que le indiquemos. 

````python
# Como hasta ahora en este jupyter trabajaremos con el dataset de Marketing que usamos en la lecci칩n de filtrado

df = pd.read_csv("Marketing-Customer-Analysis.csv")
df.head(1)
````

````python
# vamos a homogeneizar los nombres de las columnas

nuevas_columnas = {col: col.lower().replace(" ", "_") for col in df.columns}
df.rename(columns = nuevas_columnas, inplace = True)
df.head(1)
````

### Uso de apply

````python
# Lo primero que vamos a hacer es crearnos una copia de nuestro DataFrame sobre el que trabajaremos. 

df_copia = df.copy()
df_copia.head(2)
````

Para entender el `apply` en este jupyter trabajaremos con la columna de `effective_to_date`. Lo que haremos ser치 crear una funci칩n que nos separe la fecha en a침o y mes. Con esto crearemos dos columnas, una la llamaremos "a침o" y otra "mes". 

````python
def sacar_mes(fecha):
    return fecha.split("/")[0]
````

````python
#creemos la funci칩n para sacar el a침o

def sacar_a침o(x):
    '''
    Funci칩n que recibe una fecha y extrae el a침o
    
    Args: 
        x(str): cada fecha que tenemos en nuestro dataframe
    Returns: 
        el a침o en formato string
    '''
    return x.split("/")[2]

def sacar_mes(x):
    '''
    Funci칩n que recibe una fecha y extrae el mes
    
    Args: 
        x(str): cada fecha que tenemos en nuestro dataframe
    Returns: 
        el mes en formato string
    '''
    return x.split("/")[0]
````

````python
df_copia["a침o"] = (df_copia["effective_to_date"].apply(sacar_a침o))
df_copia["mes"] = (df_copia["effective_to_date"].apply(sacar_mes))
````

````python
# si nos fijamos ahora tenemos dos columnas nuevas, "a침o" y "mes"

df_copia.head(2)
````

Esta misma operaci칩n la podemos hacer con la librer칤a `datetime`. Exploremosla un poco. 

````python
# lo primero que tenemos que hacer es importarla
import datetime
````

````python
# para trabajar con esta librer칤a lo primero que tenemos que hacer es convertir el tipo de la
# columna "effective_to_date" a formato datetime. Para eso usaremos un apply

df_copia["effective_to_date"] = df_copia["effective_to_date"].apply(pd.to_datetime)
df_copia.head(1)
````

````python
# dentro de esta librer칤a tenemos much칤simas funcionalidades. Entre ellas, tenemos el m칠todo "strftime" que nos va a 
# permitir extraer el a침o, el mes o el d칤a en distintos formatos. 

def sacar_a침o2(x):
    return x.strftime("%Y")

def sacar_mes2(x):
    return x.strftime("%B")
````

````python
df_copia["a침o2"] = df_copia["effective_to_date"].apply(sacar_a침o2)
df_copia["mes2"] = df_copia["effective_to_date"].apply(sacar_mes2)
````

````python
# si nos fijamos ahora tenemos una columna nueva con el a침o (a침o2) y otra con el mes completo en letras (mes2)

df_copia.sample(1)
````

**쯈u칠 pasa si queremos usar una funci칩n que recibe dos par치metros?** 

En este caso, la cosa se va a complicar un poquito m치s, ya que tendremos que usar una `lambda` para usar el `apply`. Aqu칤 no vamos a profundizar m치s en la `lambda`, ya que nos queremos centrar en el `apply` en s칤.

Imaginemos ahora que queremos saber la media del dinero que se han llevado los clientes por cada reclamaci칩n que tienen. Para esto, tendremos en cuenta el dinero que ha recibido cada cliente en sus reclamaciones (`total_claim_amount`) y el n칰mero de reclamaciones que puso (`number_of_open_complaints`). Lo que tendremos que hacer es una divisi칩n de esas dos columnas. 

````python
def media_reclamaciones (dinero, reclamaciones):
    '''
    Esta funci칩n calcula la media de dinero que se lleva cada cliente por reclamaci칩n.
    Args: 
        dinero (int o float): el dinero que ha recibido cada cliente en sus reclamaciones
        reclamaciones (int): el n칰mero de reclamaciones que se ha puesto el cliente
    Returns: 
        Int o float con la media de dinero por reclamaci칩n. 
    '''
    # en este caso tendr칠 que hacer un try except porque puede que haya alg칰n cliente que no haya puesto ninguna reclamaci칩n
    try: 
        return dinero/reclamaciones
    except: 
        return "no puso ninguna reclamaci칩n"
````

````python
df_copia.head()
````

````python
df_copia["media_reclamaciones"] = df_copia.apply(lambda pepe: media_reclamaciones( # en este caso, no especificamos el nombre de la columna, haremos el apply sobre todo el dataframe. Dentro del apply iniciamos el lambda que recibir치 unos datos. En este caso todo el dataframe 
                                        pepe['total_claim_amount'], pepe['number_of_open_complaints']), # especificamos las columnas sobre las que queremos aplicar el apply
                                        axis = 1) # especificamos es el eje 1 ya que estamos trabajando con columnas. 
````

## Escritura secuencial de datos con bucles `for`

En este apartado lo que haremos ser치 realizar varios `groupby` con los que sacaremos la media, la desviaci칩n est치ndar y la mediana de cada uno de los grupos para las columnas de `customer_lifetime_value` y `number_of_policies`. 

Para hacer esto haremos un *for loop* y despu칠s lo almacenaremos todo en un Excel, donde cada agrupaci칩n se guarde en una hoja diferente.

````python
df_copia.columns
````

````python
# lo primero que hacemos es definir los distintos grupos que vamos a querer definir. 
grupo_clientes1 = ['policy_type','vehicle_size']
grupo_clientes2 = ['state','employmentstatus', 'coverage']

valores = ['customer_lifetime_value', 'number_of_policies' ] # las columnas sobre las que queremos aplicar las operaciones en el groupy
````

````python
df.groupby(["police_type", "state"])["customer_lifetime_value"].mean()
df.groupby(["police_type", "emplymentstatus"])["customer_lifetime_value"].mean()
````

Lo que vamos a querer es hacer muchos `groupby`s un 칰nico *for loop*, esto nos va a permitir sacar todas las combinaciones de `groupby` entre la lista de `grupo_clientes1` y `grupo_clientes2`. 

Es decir: 

* policy_type vs state
* policy_type vs employmentstatus
* policy_type vs coverage

y

* vehicle_size vs state
* vehicle_size vs employmentstatus
* vehicle_size vs coverage

Adem치s vamos a querer calcular la media, mediana y std para dos variables num칠ricas. Estas son las que est치n en la lista de `valores`. 

Visto en c칩digo querremos: 

```python
df.groupby(["police_type", "state"])["customer_lifetime_value"].mean()
df.groupby(["police_type", "emplymentstatus"])["customer_lifetime_value"].median()
df.groupby(["police_type", "emplymentstatus"])["customer_lifetime_value"].std()

df.groupby(["police_type", "state"])["number_of_policies"].mean()
df.groupby(["police_type", "state"])["number_of_policies"].median()
df.groupby(["police_type", "emplymentstatus"])["number_of_policies"].std()

# y as칤 para todas las posibles combinaciones de las listas grupo_clientes1 y grupo_clientes2

# Pero recordemos que si en un groupby queremos calcular varios estad칤sticos deberemos usar el m칠todo agg. Por lo que podr칤amos reducir un poco el c칩digo anterior a la siguiente forma 
df.groupby(["policy_type", "state"])['customer_lifetime_value'].agg(["mean", 'std', 'median'])

# si a esto le a침adimos un reset_index() nos devolver치 un dataframe
(df.groupby(["policy_type", "state"])['customer_lifetime_value'].agg(["mean", 'std', 'median'])).reset_index()
```

````python
(df.groupby(["policy_type", "state"])['customer_lifetime_value'].agg(["mean", 'std', 'median'])).reset_index()
````

````python
todos_resultados = [] # creamos una lista donde iremos guardando "todos los resultados de los groupby" que ser치n nuestros distintos dataframes

# 游뚿游뚿 Recordemos que en una lista podemos almacenar cualquier cosa, diccionarios, m치s listas, DataFrames, etc. 游뚿游뚿

print("Todas las combinaciones que nos salen de este triple for son: ") #  hacemos un print para ver todas las posibles combinaciones entre nuestros grupos
for g1 in grupo_clientes1: # iteramos por el primer grupo
    for g2 in grupo_clientes2: # iteramos por el segundo grupo 
        for val in valores: #iteramos por las columnas sobre las que queremos calcular los estad칤sticos
            print(g1, "-->", g2, '-->', val) # sacamos todas las combinaciones
            resultado = (df.groupby([g1, g2])[val].agg(["mean", 'std', 'median'])).reset_index() # hacemos el groupby y lo convertimos a dataframe
            todos_resultados.append(resultado) # apendemos los resultados del groupby en la lista vac칤a que creamos al inicio
````

````python
len(todos_resultados)
````

````python
# vemos que los elementos de la lista son cada uno de los DataFrames

todos_resultados[11]
````

Hasta aqu칤 tenemos todos los resultados del groupby almacenados en la lista `todos_resultados`. Lo siguiente que tenemos que hacer es iterar por esta lista e ir guard치ndolos en el excel. 

````python
# creamos una lista con los nombres que queremos asignar a cada una de las hojas de nuestro excel

nombres = ['policy-state-lifetime', 'policy-state-number', 
           'policy-employment-lifetime', 'policy-employment-number', 
          'policy-coverage-lifetime', 'policy-coverage-number', 
          'size-state-lifetime', 'size-state-number', 
          'size-employment-lifetime', 'size-employment-number', 
          'size-coverage-lifetime', 'size-coverage-number']
````

````python
# iteramos por la lista de resultados para acceder a cada una de las agrupaciones y poder guardarlas en el excel: 
from pandas import ExcelWriter

w = ExcelWriter("datos/Agrupaciones.xlsx") # iniciamos la creaci칩n del excel desde pandas

for resultado in range(len(todos_resultados)): # iteramos por la lista de resultados (cada iterable ser치 un DataFrame) 
    todos_resultados[resultado].to_excel(w, sheet_name=nombres[resultado]) # vamos creando tantas hojas de excel como elementos (DataFrames) tenemos en nuestra lista 
w.save() # lo guardamos todo
````
